```python
# merging and joing data frames 
df1=pd.DataFrame({'key':['A','B','C'],'value1':[1,3,5]})
df2=pd.DataFrame({'key':['A','B','D'],'value1':[134,221,451]})
```

```python
df2
```

```python
# merge dataframe on the 'key colums'
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="inner")
```
